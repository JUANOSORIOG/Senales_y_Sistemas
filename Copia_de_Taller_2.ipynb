Serie de Fourier (Exponencial, Trigonométrica y Compacta):

Semejanzas: Todas representan una señal periódica como una suma (infinita) de exponenciales complejas (exponencial) o de senos y cosenos (trigonométrica y compacta) en diferentes frecuencias.
Diferencias:
Serie de Fourier Exponencial: Utiliza exponenciales complejas $e^{j n \omega_0 t}$. Es la forma más general y compacta matemáticamente.

Serie de Fourier Trigonométrica: Utiliza senos y cosenos. Es más intuitiva para visualizar las componentes de frecuencia reales.

Serie de Fourier Compacta: Es una variación de la trigonométrica que combina los términos de seno y coseno en una sola forma con amplitud y fase.

Se aplica a señales periódicas y continuas en el tiempo. El resultado es un espectro de frecuencias discreto (solo existen componentes en las frecuencias fundamentales y sus armónicos).


Transformada Discreta de Fourier (DFT):

Semejanzas: Es una versión discreta de la DTFT. También descompone una señal discreta en componentes de frecuencia.

Diferencias: Se aplica a señales discretas y de duración finita. El resultado es un espectro de frecuencias discreto. La DFT se puede calcular eficientemente mediante el algoritmo de la Transformada Rápida de Fourier (FFT).

Se aplica en Análisis de señales discretas de duración finita, procesamiento digital de señales, implementación computacional de la transformada de Fourier.


La relación entre el dominio del tiempo y el dominio de la frecuencia (espectro) se rige por las transformadas de Fourier. Aquí están las cuatro combinaciones principales:

Señal en Tiempo Continuo - Espectro Continuo:

Descripción: Señales que existen para todos los valores de tiempo y su contenido de frecuencia también está presente en un rango continuo de frecuencias.

**Transformada asociada**: Transformada de Fourier continua.
Ejemplo: Una onda sinusoidal pura (en teoría, una señal infinita) tiene un espectro que es un impulso en una frecuencia única. Una señal de audio analógica es un ejemplo más práctico de una señal continua en el tiempo con un espectro continuo.

**Señal en Tiempo Continuo - Espectro Discreto**:
Descripción: Señales que existen para todos los valores de tiempo, pero su contenido de frecuencia solo está presente en frecuencias específicas (múltiplos de una frecuencia fundamental).

**Transformada asociada**: Serie de Fourier.
Ejemplo: Una señal periódica continua (como una onda cuadrada o triangular ideal) puede representarse como una suma de sinusoides en frecuencias discretas (la frecuencia fundamental y sus armónicos).

**Señal en Tiempo Discreto - Espectro Continuo y Periódico**:
Descripción: Señales que solo existen en instantes de tiempo discretos, pero su contenido de frecuencia está presente en un rango continuo de frecuencias, repitiéndose periódicamente.

**Transformada asociada**: Transformada de Fourier en Tiempo Discreto (DTFT).
Ejemplo: Una señal continua que ha sido muestreada da como resultado una señal en tiempo discreto. Su espectro es continuo y periódica debido al proceso de muestreo (aliasing si no se cumple el teorema de Nyquist).

**Señal en Tiempo Discreto - Espectro Discreto**:
Descripción: Señales que solo existen en instantes de tiempo discretos y tienen una duración finita. Su contenido de frecuencia también está presente solo en frecuencias discretas.

Transformada asociada**: Transformada Discreta de Fourier (DFT).
Ejemplo: Una secuencia finita de muestras de una señal, como un fragmento de audio digital o una imagen digital. La DFT se utiliza para analizar el contenido de frecuencia de estas señales finitas.

# ***Algoritmo Fast Fourier Transform (FFT):***

# **¿En qué consiste y su utilidad para el cálculo de la DFT?**

La Fast Fourier Transform (FFT) no es una transformada diferente, sino un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT). La DFT es fundamental en el procesamiento digital de señales, pero su cálculo directo puede ser computacionalmente muy costoso, especialmente para señales largas. La FFT reduce drásticamente el número de operaciones necesarias, haciendo que el análisis y procesamiento de señales digitales a gran escala sea práctico.

Su utilidad radica en permitir el cálculo rápido de la DFT, lo cual es esencial en una amplia gama de aplicaciones, incluyendo:

Análisis de espectro de audio y video.

Compresión de datos (JPEG, MP3).

Filtrado digital.

Modulación y demodulación en sistemas de comunicación.

Resolución de ecuaciones diferenciales parciales.

Análisis de vibraciones y sistemas mecánicos.

**Explicación detallada del algoritmo FFT**:

Existen varios algoritmos FFT, pero el más común es el algoritmo de Cooley-Tukey. Este algoritmo se basa en el principio de "divide y vencerás". La idea principal es descomponer una DFT de tamaño N en dos DFTs de tamaño N/2. Este proceso se repite recursivamente hasta que se obtienen DFTs de tamaño muy pequeño (generalmente de tamaño 2), cuyo cálculo es trivial.

Para que el algoritmo de Cooley-Tukey sea más eficiente, N generalmente debe ser una potencia de 2 (por ejemplo, 256, 512, 1024, etc.). Si N no es una potencia de 2, se puede "rellenar con ceros" la señal hasta la siguiente potencia de 2.

El proceso general para un algoritmo FFT de radix-2 (donde N es una potencia de 2) implica los siguientes pasos (simplificado):

División: La secuencia de entrada de longitud N se divide en dos subsecuencias: una con los elementos de índice par y otra con los elementos de índice impar.
Recursión: Se calcula la DFT de cada una de estas subsecuencias de longitud N/2 utilizando el mismo algoritmo FFT.

**Combinación**: Los resultados de las dos DFTs de tamaño N/2 se combinan para obtener la DFT de tamaño N. Esta combinación implica operaciones de multiplicación por "factores de giro" (twiddle factors), que son exponenciales complejas, y sumas.

Este proceso recursivo de división y combinación reduce significativamente el número de operaciones.

# ***Costo Computacional: DFT vs. FFT***

El costo computacional se mide generalmente por el número de multiplicaciones y sumas complejas requeridas.

DFT directa:

El cálculo directo de la DFT de una secuencia de longitud N requiere aproximadamente $N^2$$N^2$ multiplicaciones complejas y $N(N-1)$$N(N-1)$ sumas complejas.
El costo computacional es del orden de $O(N^2)$$O(N^2)$. Para señales largas, este número crece muy rápidamente, haciendo el cálculo inviable.

**FFT (radix-2):**

El algoritmo FFT de radix-2 para una secuencia de longitud N (donde N es una potencia de 2) requiere aproximadamente $(N/2) \log_2(N)$$(N/2) \log_2(N)$ multiplicaciones complejas y $N \log_2(N)$$N \log_2(N)$ sumas complejas.

El costo computacional es del orden de $O(N \log_2(N))$$O(N \log_2(N))$.

Comparación del Costo Computacional:

La diferencia entre $O(N^2)$$O(N^2)$ y $O(N \log_2(N))$$O(N \log_2(N))$ es enorme para valores grandes de N.

Por ejemplo, si $N = 1024 = 2^{10}$$N = 1024 = 2^{10}$:

DFT directa:

$1024^2 \approx 1,000,000$$1024^2 \approx 1,000,000$ operaciones.
FFT: $1024 \times \log_2(1024) = 1024 \times 10 \approx 10,000$$1024 \times \log_2(1024) = 1024 \times 10 \approx 10,000$ operaciones.

En este ejemplo, la FFT es aproximadamente 100 veces más rápida que la DFT directa. A medida que N aumenta, la ventaja de la FFT se vuelve aún más dramática.

# **Modulación por Amplitud (AM)**

La Modulación por Amplitud (AM) es una técnica en la que la amplitud de una señal portadora de alta frecuencia varía en proporción a la amplitud de una señal moduladora (señal de información). La expresión matemática de una señal AM es:

$s(t) = A_c [1 + m(t)] \cos(2\pi f_c t)$

Donde:

*   $A_c$: Amplitud de la portadora.
*   $m(t)$: Señal moduladora (normalizada).
*   $f_c$: Frecuencia de la portadora.

# **Detección Coherente (Demodulación AM)**

La detección coherente es un método de demodulación que requiere una réplica exacta de la portadora original en frecuencia y fase para recuperar la señal original. El proceso es el siguiente:

1.  **Multiplicación de la señal AM recibida por una portadora local coherente:**

    $r(t) = s(t) \cdot \cos(2\pi f_c t + \phi)$
    donde $\phi$ es la diferencia de fase entre la portadora local y la original.

2.  **Filtrado paso bajo (LPF):** para eliminar las componentes de alta frecuencia y recuperar la señal original $m(t)$.

Si $\phi = 0$ (sin desfase), la demodulación es óptima. Si $\phi \neq 0$, puede haber distorsión o atenuación.

# ***Señal mensaje tipo coseno:***

```python
import numpy as np
import matplotlib.pyplot as plt
#Frecuencia portadora y muestreo

Fc = 100 #frecuencia de portadora

Fm = 10 #frecuencia del mensaje
Am = 1 #amplitud del mensaje


ti = 0
tf = 5/Fm #cantidad de periodos a graficar de la

Fs = 5*Fc #frecuencia muestreo

t = np.arange(ti,tf,1/Fs) #vector del tiempo
m = Am*np.cos(2*np.pi*Fm*t) #mensaje #

Im = 0.8 # Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im

Ac = max(abs(m))/Im #amplitud de la portadora según índice de modulación
c = Ac*np.cos(2*np.pi*Fc*t) #portadora
print(Ac)
```

```python
plt.plot(t,c,label='portadora')
plt.plot(t,m,label='mensaje')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Frecuencia portadora y muestreo

Fc = 100 #frecuencia de portadora

Fm = 10 #frecuencia del mensaje
Am = 1 #amplitud del mensaje


ti = 0
tf = 5/Fm #cantidad de periodos a graficar de la

Fs = 5*Fc #frecuencia muestreo

t = np.arange(ti,tf,1/Fs) #vector del tiempo

# Define a rectangular pulse message
m_pulse = np.zeros_like(t)
pulse_start = 2/Fm
pulse_end = 3/Fm
m_pulse[(t >= pulse_start) & (t <= pulse_end)] = Am

Im = 1 # Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im

Ac = max(abs(m_pulse))/Im #amplitud de la portadora según índice de modulación
c = Ac*np.cos(2*np.pi*Fc*t) #portadora
print(Ac)

plt.plot(t,c,label='portadora')
plt.plot(t,m_pulse,label='mensaje pulso rectangular')
plt.legend()
plt.show()
```

# **Aplicación en Circuitos Eléctricos: Potencia y Distorsión Armónica**

## **Distorsión Total de Armónicos (Total Harmonic Distortion - THD)**

La Distorsión Total de Armónicos (THD) es una medida de cuán distorsionada está una forma de onda de una señal periódica en comparación con una onda sinusoidal pura. En sistemas de potencia eléctrica, una forma de onda de tensión o corriente ideal es sinusoidal a la frecuencia fundamental (por ejemplo, 50 Hz o 60 Hz). Sin embargo, debido a cargas no lineales (como rectificadores, inversores, fuentes de alimentación conmutadas, etc.), las formas de onda pueden volverse distorsionadas, conteniendo componentes en frecuencias que son múltiplos enteros de la frecuencia fundamental. Estas componentes son los armónicos.

El THD se define como la relación entre la potencia (o RMS) de todos los componentes armónicos y la potencia (o RMS) del componente fundamental. Se expresa generalmente como un porcentaje.

Para una tensión $V(t)$:
$$THD_V = \frac{\sqrt{V_2^2 + V_3^2 + V_4^2 + ...}}{V_1} \times 100\%$$
Para una corriente $I(t)$:
$$THD_I = \frac{\sqrt{I_2^2 + I_3^2 + I_4^2 + ...}}{I_1} \times 100\%$$
Donde $V_n$ e $I_n$ son los valores RMS de la n-ésima componente armónica de tensión y corriente, respectivamente, y $V_1$ e $I_1$ son los valores RMS del componente fundamental.

Un THD bajo indica que la forma de onda se parece mucho a una sinusoide pura. Un THD alto indica una distorsión significativa. La distorsión armónica puede causar varios problemas en los sistemas de potencia, como sobrecalentamiento de transformadores y motores, resonancia, disparo intempestivo de protecciones, y errores en equipos de medición.

## **Factor de Potencia**

El factor de potencia (FP) en un circuito eléctrico es una medida de cuán eficientemente se está utilizando la energía eléctrica. Se define como la relación entre la potencia real (o activa) consumida por la carga y la potencia aparente suministrada por la fuente.

$$FP = \frac{\text{Potencia Real (P)}}{\text{Potencia Aparente (S)}}$$

En circuitos con formas de onda puramente sinusoidales, el factor de potencia es simplemente el coseno del ángulo de desfase entre la tensión y la corriente ($FP = \cos(\theta)$). Esto se conoce como factor de potencia de desplazamiento.

Sin embargo, en presencia de distorsión armónica (formas de onda no sinusoidales), el factor de potencia se ve afectado tanto por el desfase entre los componentes fundamentales de tensión y corriente como por la presencia de los armónicos. En este caso, el factor de potencia total se define como:

$$FP = \frac{P}{V_{RMS\_total} \times I_{RMS\_total}}$$

Donde $V_{RMS\_total}$ y $I_{RMS\_total}$ son los valores RMS totales de la tensión y la corriente, respectivamente, incluyendo todos los armónicos.

Un factor de potencia ideal es 1. Un factor de potencia bajo implica que se está suministrando más potencia aparente de la que realmente se consume como potencia útil (real), lo que lleva a mayores pérdidas en la red eléctrica y penalizaciones por parte de las compañías eléctricas.

## **Cálculo de THD desde la FFT**

La Transformada Rápida de Fourier (FFT) es una herramienta fundamental para calcular el THD. La FFT descompone una señal en el dominio del tiempo en sus componentes de frecuencia (su espectro). Los picos en el espectro corresponden a las amplitudes (o potencias) de las diferentes componentes de frecuencia presentes en la señal, incluyendo el fundamental y sus armónicos.

El proceso para calcular el THD de una señal (tensión o corriente) usando la FFT es el siguiente:

1. **Adquirir la señal:** Se toma una muestra digital de la forma de onda de tensión o corriente de interés durante un período (o varios períodos) utilizando un sistema de adquisición de datos a una frecuencia de muestreo suficientemente alta para capturar los armónicos relevantes (de acuerdo con el teorema de Nyquist).

2. **Aplicar la FFT:** Se aplica el algoritmo FFT a la secuencia de muestras digitales obtenida. Esto produce un espectro de frecuencia discreto, donde cada "bin" de frecuencia corresponde a una componente de frecuencia específica.

3. **Identificar las componentes:** Se identifica el pico correspondiente a la frecuencia fundamental ($f_1$) y los picos correspondientes a sus armónicos ($f_2=2f_1, f_3=3f_1$, etc.).

4. **Calcular las amplitudes (o RMS):** A partir de la salida de la FFT, se pueden calcular las amplitudes (o más comúnmente, los valores RMS) de la componente fundamental ($V_1$ o $I_1$) y de cada componente armónica ($V_n$ o $I_n$). Es importante tener en cuenta la escala de la FFT y si la salida representa amplitudes pico, RMS, o densidad espectral.

5. **Calcular el THD:** Se utiliza la fórmula del THD mencionada anteriormente, sumando cuadráticamente los valores RMS de los armónicos y dividiendo por el valor RMS del fundamental.

$$THD = \frac{\sqrt{\sum_{n=2}^{N} V_n^2}}{V_1} \times 100\%$$
ó
$$THD = \frac{\sqrt{\sum_{n=2}^{N} I_n^2}}{I_1} \times 100\%$$
Donde N es el número máximo de armónicos considerados (generalmente hasta el armónico 25, 50 o según la normativa aplicable).

En resumen, la FFT proporciona las amplitudes de las diferentes componentes de frecuencia presentes en una señal, lo que permite cuantificar la contribución de los armónicos y, por lo tanto, calcular el THD.

## **Resumen: Distorsión del Factor de Potencia y THD**

El factor de potencia total ($FP_{total}$) en circuitos con formas de onda distorsionadas se ve afectado por dos factores:

1.  **Factor de Potencia de Desplazamiento ($FP_{desplazamiento}$):** Relacionado con el desfase entre las componentes fundamentales de tensión y corriente ($\cos(\theta_1)$).
2.  **Factor de Potencia de Distorsión ($FP_{distorsión}$):** Relacionado con la presencia de armónicos (cuantificados por el THD).

La relación es: $FP_{total} = FP_{desplazamiento} \times FP_{distorsión}$.

El Factor de Potencia de Distorsión ($FP_{distorsión}$) se calcula a partir del THD de la corriente (asumiendo tensión sinusoidal) con la fórmula:

$$FP_{distorsión} = \frac{1}{\sqrt{1 + (\text{THD}_I/100)^2}}$$

**En resumen:** El THD cuantifica la distorsión de la forma de onda. Esta distorsión *reduce* el factor de potencia total a través del $FP_{distorsión}$. Un THD alto implica un $FP_{distorsión}$ bajo, lo que disminuye el $FP_{total}$, incluso si el $FP_{desplazamiento}$ es bueno.
